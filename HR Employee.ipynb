Data Preprocessing

```python
# Coverting data with type object to numeric with labelencoder
from sklearn.preprocessing import LabelEncoder
g=df_hr.columns
for i in g:
    l=LabelEncoder()
    if df_hr[i].dtypes=="object":
        df_hr[i]=l.fit_transform(df_hr[i])
        
df_hr.dtypes
```

```python
#detectiong outlier with z score
from scipy.stats import zscore #importing the zscore
z=np.abs(zscore(df_hr))
df_hr_final=df_hr[(z<3).all(axis=1)]#makinhg the finala data frame
print(df_hr.shape,df_hr_final.shape)
```

Data Standardize

```python
#Seprating input & output variable
df_x=df_hr_final.drop(columns=["Attrition"])
y=df_hr_final[["Attrition"]]
print(df_x.shape,y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(df_x,columns=df_x.columns)
```

Model Creating



```python
#Creating a function for finding the better random state for each model
#As we know that our target variable is Attrition, which is catagorical so we are using here the LOGISTIC regression
#We are trying to find the best random satate for getting the better accuracy score

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
from sklearn.model_selection import train_test_split
def hr_acc(clf,x,y):
    max_acc=0
    for rd in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=rd,test_size=0.20)
        clf.fit(x_train,y_train)
        y_prd=clf.predict(x_test)
        asc=accuracy_score(y_test,y_prd)
        if asc>max_acc:
            max_acc=asc
            final_rd=rd
    print("max accuracy score coressponding to  Random_state→",final_rd,"is",max_acc*100)
    return final_rd
```

```python
#Making function for confusion matrix & classification report
def cc(md,x,y,rd):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=rd,test_size=0.20)
    md.fit(x_train,y_train)
    pre=md.predict(x_test)
    acc=accuracy_score(y_test,pre)
    print(acc)
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test,pre)
    roc_auc = auc(false_positive_rate, true_positive_rate)
    print ('roc_auc_score = ',roc_auc)
    cm=confusion_matrix(y_test,pre)
    print()
    print()
    cr=classification_report(y_test,pre)
    print()
    print()
    print()
    print(cr,"\n","☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼☼")
    plt.figure(figsize=[20,50],facecolor='green')
    plt.subplot(912)
    plt.title(md,{"fontsize":22})
    plt.plot(false_positive_rate, true_positive_rate, label='AUC = %0.2f'% roc_auc)
    plt.plot([0,1],[0,1],'r--')
    plt.legend(loc='lower right')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

```python
#Function from cross_val_score
from sklearn.model_selection import cross_val_score   
def cvs(model,x,y):
    c=cross_val_score(model,x,y,cv=5,scoring="accuracy")
    print("mean accuracy  score for ",model,c.mean())
    print("Standard deviation  in accuracy score for ",model,c.std())
    print()
    print("******************************************************")
    print("After seen the cross validation score of",model,"the accuracy score mean is",c.mean())
```

```python
#Function for storing the result
def rslt(model,accuracy,n):
    D=pd.DataFrame({"MODEL":[model],"ACCURACY":[accuracy]},index={n})
    return D
```
